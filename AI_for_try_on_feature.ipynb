```python
def run_tryon(user_img_path, product_img_path, output_dir="/content/drive/MyDrive/Colab Notebooks/Try on feature/output"):
    # Enhance image if low quality
    if is_low_quality(user_img_path):
        !python Real-ESRGAN/inference_realesrgan.py -i {user_img_path} -o {os.path.dirname(user_img_path)} -n RealESRGAN_x4plus
        enhanced_path = os.path.join(os.path.dirname(user_img_path), os.path.basename(user_img_path).replace(".jpg", "_out.jpg"))
        if os.path.exists(enhanced_path):
            os.replace(enhanced_path, user_img_path)
    # Prepare VITON-HD input
    os.makedirs("VITON-HD/datasets/test/image", exist_ok=True)
    os.makedirs("VITON-HD/datasets/test/cloth", exist_ok=True)
    !cp {user_img_path} VITON-HD/datasets/test/image/
    !cp {product_img_path} VITON-HD/datasets/test/cloth/
    !echo "$(basename {user_img_path}) $(basename {product_img_path})" > VITON-HD/datasets/test_pairs.txt
    # Run try-on
    %cd VITON-HD
    !CUDA_VISIBLE_DEVICES=0 python test.py --name VITONHD --dataroot ./datasets/test --save_dir {output_dir}
    %cd ..
    # Return output path
    output_path = f"{output_dir}/try-on/$(basename {user_img_path})"
    return output_path if os.path.exists(output_path) else None

# Example usage:
# output = run_tryon(user_img_path, product_img_path)
```

```python
from fastapi import FastAPI, UploadFile, File
import shutil

app = FastAPI()

@app.post("/tryon")
async def tryon(user_image: UploadFile = File(...), product_image: UploadFile = File(...)):
    user_path = "/content/drive/MyDrive/Colab Notebooks/Try on feature/user_input/user_image.jpg"
    product_path = "/content/drive/MyDrive/Colab Notebooks/Try on feature/product_input/product_image.jpg"
    with open(user_path, "wb") as buffer:
        shutil.copyfileobj(user_image.file, buffer)
    with open(product_path, "wb") as buffer:
        shutil.copyfileobj(product_image.file, buffer)
    output = run_tryon(user_path, product_path)
    return {"output_path": output}
```
